```python
#=====SDSS Datathon - Public Services (Shelter Occupancy)=====

import pandas as pd
import numpy as np
import re

#-----File path + load data-----
FILEPATH = "public_services_dataset.xlsx"  # our data path

#There are 2 sheets in the excel: 'data' and 'dictionary'
df = pd.read_excel(FILEPATH, sheet_name="dataset")
data_dict = pd.read_excel(FILEPATH, sheet_name="dictionary")

print("Raw shape:", df.shape)
print("Columns:", df.columns.tolist())

#-----Standardized Column Names-----
# Keep original names but remove weird spaces
df.columns = [c.strip() for c in df.columns]

EXPECTED_COLS = [
    "OCCUPANCY_DATE",
    "LOCATION_POSTAL_CODE",
    "SECTOR",
    "OVERNIGHT_SERVICE_TYPE",
    "PROGRAM_MODEL",
    "PROGRAM_AREA",
    "CAPACITY_TYPE",
    "ACTUAL_CAPACITY",
    "OCCUPIED_CAPACITY",
    "UNAVAILABLE_CAPACITY",
    "OCCUPANCY_RATE",
]
#making sure that all important columns exist
missing = [c for c in EXPECTED_COLS if c not in df.columns]
if missing:
    raise ValueError(f"Missing expected columns: {missing}")

#-----Clean date column-----
df["OCCUPANCY_DATE"] = pd.to_datetime(df["OCCUPANCY_DATE"], errors="coerce")

# Drop rows with invalid dates (not in form YYYY-MM-DD)
bad_dates = df["OCCUPANCY_DATE"].isna().sum()
if bad_dates:
    print(f"Warning: {bad_dates} rows have invalid OCCUPANCY_DATE -> dropping them.")
    df = df.dropna(subset=["OCCUPANCY_DATE"]).copy()

#-----Clean categorical/text columns-----
CAT_COLS = [
    "LOCATION_POSTAL_CODE",
    "SECTOR",
    "OVERNIGHT_SERVICE_TYPE",
    "PROGRAM_MODEL",
    "PROGRAM_AREA",
    "CAPACITY_TYPE",
]

for c in CAT_COLS:
    df[c] = df[c].astype("string").str.strip()

#Postal code standardization (uppercase andremove spaces)
df["LOCATION_POSTAL_CODE"] = (
    df["LOCATION_POSTAL_CODE"]
    .str.upper()
    .str.replace(r"\s+", "", regex=True)
)

#Basic postal code validity flag (not a hard filter)
postal_pattern = re.compile(r"^[A-Z]\d[A-Z]\d[A-Z]\d$")
df["POSTAL_VALID_FLAG"] = df["LOCATION_POSTAL_CODE"].apply(
    lambda x: bool(postal_pattern.match(x)) if pd.notna(x) else False
)

# -----Convert numeric columns-----

NUM_COLS = ["ACTUAL_CAPACITY", "OCCUPIED_CAPACITY", "UNAVAILABLE_CAPACITY", "OCCUPANCY_RATE"]

for c in NUM_COLS:
    df[c] = pd.to_numeric(df[c], errors="coerce")

#If critical numeric columns missing, drop those rows
critical_missing = df[["ACTUAL_CAPACITY", "OCCUPIED_CAPACITY"]].isna().any(axis=1).sum()
if critical_missing:
    print(f"Warning: {critical_missing} rows missing ACTUAL/OCCUPIED capacity -> dropping them.")
    df = df.dropna(subset=["ACTUAL_CAPACITY", "OCCUPIED_CAPACITY"]).copy()

#Fill with 0 for the datasets UNAVAILABLE_CAPACITY that missing
df["UNAVAILABLE_CAPACITY"] = df["UNAVAILABLE_CAPACITY"].fillna(0)

#Ensure capacities are not negative
for c in ["ACTUAL_CAPACITY", "OCCUPIED_CAPACITY", "UNAVAILABLE_CAPACITY"]:
    df[f"{c}_NEG_FLAG"] = df[c] < 0
    df[c] = df[c].clip(lower=0)

#Convert capacities to integers if they are whole numbers
for c in ["ACTUAL_CAPACITY", "OCCUPIED_CAPACITY", "UNAVAILABLE_CAPACITY"]:
    # keep as Int64 to preserve missing (though we filled most)
    df[c] = df[c].round().astype("Int64")

#-----Derived variables-----
# Available capacity
df["AVAILABLE_CAPACITY"] = (df["ACTUAL_CAPACITY"] - df["UNAVAILABLE_CAPACITY"]).astype("Int64")

#Available capacity should not be negative
df["AVAILABLE_CAPACITY_NEG_FLAG"] = df["AVAILABLE_CAPACITY"] < 0
df["AVAILABLE_CAPACITY"] = df["AVAILABLE_CAPACITY"].clip(lower=0).astype("Int64")

#Occupancy rate computed from ACTUAL_CAPACITY
df["OCC_RATE_CALC"] = np.where(
    df["ACTUAL_CAPACITY"].astype(float) > 0,
    df["OCCUPIED_CAPACITY"].astype(float) / df["ACTUAL_CAPACITY"].astype(float),
    np.nan
)

#Occupancy rate computed from AVAILABLE_CAPACITY
df["OCC_RATE_AVAILABLE"] = np.where(
    df["AVAILABLE_CAPACITY"].astype(float) > 0,
    df["OCCUPIED_CAPACITY"].astype(float) / df["AVAILABLE_CAPACITY"].astype(float),
    np.nan
)

#Capacity loss ratio
df["CAPACITY_LOSS_RATIO"] = np.where(
    df["ACTUAL_CAPACITY"].astype(float) > 0,
    df["UNAVAILABLE_CAPACITY"].astype(float) / df["ACTUAL_CAPACITY"].astype(float),
    np.nan
)

#Sanity flags for rate mismatch (provided vs calculated)
df["OCC_RATE_MISMATCH_FLAG"] = np.where(
    df["OCCUPANCY_RATE"].notna(),
    (np.abs(df["OCCUPANCY_RATE"].astype(float) - df["OCC_RATE_CALC"].astype(float)) > 1e-6),
    False
)

# Critical thresholds
df["CRITICAL_95_FLAG"] = df["OCC_RATE_CALC"] >= 0.95
df["OVERFLOW_FLAG"] = df["OCCUPIED_CAPACITY"] > df["ACTUAL_CAPACITY"]
df["OVERFLOW_AVAILABLE_FLAG"] = df["OCCUPIED_CAPACITY"] > df["AVAILABLE_CAPACITY"]

#-----Build a stable SHELTER_ID-----
#No explicit program_id, so we create a consistent identifier from stable attributes
#include everything that defines a "program observed daily"
ID_COLS = [
    "LOCATION_POSTAL_CODE",
    "SECTOR",
    "OVERNIGHT_SERVICE_TYPE",
    "PROGRAM_MODEL",
    "PROGRAM_AREA",
    "CAPACITY_TYPE",
]

df["SHELTER_ID"] = df[ID_COLS].astype("string").fillna("").agg("|".join, axis=1)

#-----De-duplicate exact duplicates-----
#If there are exact duplicates (same day, same shelter_id, same numbers), drop them
before = df.shape[0]
df = df.drop_duplicates(
    subset=["OCCUPANCY_DATE", "SHELTER_ID", "ACTUAL_CAPACITY", "OCCUPIED_CAPACITY", "UNAVAILABLE_CAPACITY"]
).copy()
after = df.shape[0]
if after != before:
    print(f"Dropped exact duplicates: {before - after}")

# -----Cleaning report-----
print("\nCleaned shape:", df.shape)
print("Date range:", df["OCCUPANCY_DATE"].min(), "to", df["OCCUPANCY_DATE"].max())

report = {
    "rows": int(df.shape[0]),
    "unique_shelter_ids": int(df["SHELTER_ID"].nunique()),
    "postal_invalid_rate": float((~df["POSTAL_VALID_FLAG"]).mean()),
    "occ_rate_mismatch_rate": float(df["OCC_RATE_MISMATCH_FLAG"].mean()),
    "overflow_rate_actual": float(df["OVERFLOW_FLAG"].mean()),
    "critical_95_rate": float(df["CRITICAL_95_FLAG"].mean()),
    "available_capacity_negative_rate": float(df["AVAILABLE_CAPACITY_NEG_FLAG"].mean()),
}

print("\n--- Cleaning Report ---")
for k, v in report.items():
    print(f"{k}: {v}")

#-----Save cleaned outputs-----
# Save as parquet, and csv
df.to_parquet("cleaned_shelter_occupancy.parquet", index=False)
df.to_csv("cleaned_shelter_occupancy.csv", index=False)

print("\nSaved:")
print("- cleaned_shelter_occupancy.parquet")
print("- cleaned_shelter_occupancy.csv")
```

```python
#=====FRAGILITY INDEX + SCENARIO SIMULATION + BUFFER POLICY====

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.dpi"] = 120
os.makedirs("figures", exist_ok=True)
os.makedirs("outputs", exist_ok=True)

#-----Build Shelter-Level Metrics (one row per SHELTER_ID)-----
# bcs we need stable program-level stats to rank "fragility".
def q95(x):
    x = pd.to_numeric(x, errors="coerce")
    x = x.dropna()
    return np.percentile(x, 95) if len(x) else np.nan

shelter_metrics = (
    df.groupby("SHELTER_ID")
      .agg(
          days_observed=("OCCUPANCY_DATE", "count"),
          mean_occ=("OCC_RATE_CALC", "mean"),
          p95_occ=("OCC_RATE_CALC", q95),
          max_occ=("OCC_RATE_CALC", "max"),
          std_occ=("OCC_RATE_CALC", "std"),
          mean_caploss=("CAPACITY_LOSS_RATIO", "mean"),
          critical_share=("CRITICAL_95_FLAG", "mean"),
          overflow_share=("OVERFLOW_FLAG", "mean"),
          mean_actual_capacity=("ACTUAL_CAPACITY", "mean"),
      )
      .reset_index()
)

#std can be NaN if only 1 day observed -> set to 0
shelter_metrics["std_occ"] = shelter_metrics["std_occ"].fillna(0)

print("Shelter metrics shape:", shelter_metrics.shape)
print("Example rows:")
print(shelter_metrics.head(3))

#Filter out shelters with too few days
MIN_DAYS = 30
shelter_metrics_f = shelter_metrics[shelter_metrics["days_observed"] >= MIN_DAYS].copy()
print(f"\nFiltered to shelters with >= {MIN_DAYS} days: {len(shelter_metrics_f)} shelters")

#-----Normalize Metrics to 0–1-----
#bcs each metric has different scale. Normalize so weights are fair.
def minmax(series: pd.Series) -> pd.Series:
    s = series.astype(float).to_numpy()
    mn = np.nanmin(s)
    mx = np.nanmax(s)
    if np.isclose(mx - mn, 0) or np.isnan(mn) or np.isnan(mx):
        return pd.Series(np.zeros(len(series)), index=series.index)
    return (series.astype(float) - mn) / (mx - mn)

for col in ["mean_occ", "p95_occ", "std_occ", "mean_caploss"]:
    shelter_metrics_f[col + "_n"] = minmax(shelter_metrics_f[col])

#-----Fragility Index-----
#for unique differentiator — composite score measuring vulnerability.

#Interpretable weights
W_MEAN = 0.40   # sustained pressure
W_P95  = 0.25   # peak stress events
W_STD  = 0.20   # fragility
W_CAPL = 0.15   # unavailable capacity (operational loss)

shelter_metrics_f["fragility_index"] = (
    W_MEAN * shelter_metrics_f["mean_occ_n"] +
    W_P95  * shelter_metrics_f["p95_occ_n"] +
    W_STD  * shelter_metrics_f["std_occ_n"] +
    W_CAPL * shelter_metrics_f["mean_caploss_n"]
)

#Risk tiers (relative quantiles)
p90 = shelter_metrics_f["fragility_index"].quantile(0.90)
p70 = shelter_metrics_f["fragility_index"].quantile(0.70)

shelter_metrics_f["risk_tier"] = np.select(
    [
        shelter_metrics_f["fragility_index"] >= p90,
        shelter_metrics_f["fragility_index"] >= p70
    ],
    ["Critical", "High"],
    default="Moderate/Low"
)

#Rank (1 = most fragile)
shelter_metrics_f["fragility_rank"] = shelter_metrics_f["fragility_index"].rank(
    ascending=False, method="dense"
).astype(int)

print("\nTop 10 fragile shelters:")
print(
    shelter_metrics_f.sort_values("fragility_index", ascending=False)
                    .head(10)[["SHELTER_ID","days_observed","mean_occ","p95_occ","std_occ","mean_caploss","fragility_index","risk_tier"]]
)

#Save the shelter-level index table (for slides / appendix)
shelter_metrics_f.to_csv("outputs/shelter_fragility_index.csv", index=False)
print("\nSaved: outputs/shelter_fragility_index.csv")

#-----Plot Fragility Results (Top 15)-----
#visualization of the most vulnerable programs.
topN = shelter_metrics_f.sort_values("fragility_index", ascending=False).head(15)

plt.figure(figsize=(10, 7))
plt.barh(topN["SHELTER_ID"], topN["fragility_index"])
plt.gca().invert_yaxis()
plt.title("Top 15 Most Fragile Shelter Programs (Fragility Index)")
plt.xlabel("Fragility Index (0–1 composite)")
plt.ylabel("SHELTER_ID")
plt.tight_layout()
plt.savefig("figures/model_01_fragility_top15.png", dpi=300)
plt.show()

# Distribution of fragility
plt.figure(figsize=(8, 5))
plt.hist(shelter_metrics_f["fragility_index"], bins=30)
plt.title("Distribution of Fragility Index (Shelter Programs)")
plt.xlabel("Fragility Index")
plt.ylabel("Number of Shelter Programs")
plt.tight_layout()
plt.savefig("figures/model_02_fragility_distribution.png", dpi=300)
plt.show()

#-----Scenario Simulation Engine-----
#measure resilience under shocks.
def simulate(df_in: pd.DataFrame, demand_increase=0.0, capacity_buffer=0.0) -> pd.DataFrame:
    """
    demand_increase: increases OCCUPIED_CAPACITY by % (e.g., 0.05 = +5%)
    capacity_buffer: increases ACTUAL_CAPACITY by % (e.g., 0.10 = +10%)
    Returns a copy with scenario columns.
    """
    d = df_in.copy()

    occ = d["OCCUPIED_CAPACITY"].astype(float) * (1.0 + demand_increase)
    cap = d["ACTUAL_CAPACITY"].astype(float) * (1.0 + capacity_buffer)

    occ_rate = np.where(cap > 0, occ / cap, np.nan)

    d["SCEN_OCCUPIED"] = occ
    d["SCEN_CAPACITY"] = cap
    d["SCEN_OCC_RATE"] = occ_rate
    d["SCEN_CRITICAL_95"] = occ_rate >= 0.95
    d["SCEN_OVERFLOW"] = occ > cap

    return d

def summarize_system(d: pd.DataFrame, label: str) -> dict:
    return {
        "scenario": label,
        "records": len(d),
        "avg_occ_rate": float(np.nanmean(d["SCEN_OCC_RATE"])),
        "critical_share": float(np.mean(d["SCEN_CRITICAL_95"])),
        "overflow_share": float(np.mean(d["SCEN_OVERFLOW"])),
    }

# Define scenarios
SCENARIOS = [
    ("Baseline", 0.00, 0.00),
    ("Demand +5%", 0.05, 0.00),
    ("Demand +10%", 0.10, 0.00),
    ("Capacity +5%", 0.00, 0.05),
    ("Capacity +10%", 0.00, 0.10),
    ("Demand +5% & Capacity +5%", 0.05, 0.05),
    ("Demand +10% & Capacity +10%", 0.10, 0.10),
]

system_rows = []
sector_rows = []
model_rows = []

for label, dincr, cbuf in SCENARIOS:
    dsc = simulate(df, demand_increase=dincr, capacity_buffer=cbuf)

    # System-level summary
    system_rows.append(summarize_system(dsc, label))

    # Sector breakdown
    sector_sum = (
        dsc.groupby("SECTOR")
           .agg(
               avg_occ=("SCEN_OCC_RATE", "mean"),
               critical_share=("SCEN_CRITICAL_95", "mean"),
               overflow_share=("SCEN_OVERFLOW", "mean"),
               records=("SCEN_OCC_RATE", "count"),
           )
           .reset_index()
    )
    sector_sum["scenario"] = label
    sector_rows.append(sector_sum)

    # Program model breakdown
    model_sum = (
        dsc.groupby("PROGRAM_MODEL")
           .agg(
               avg_occ=("SCEN_OCC_RATE", "mean"),
               critical_share=("SCEN_CRITICAL_95", "mean"),
               overflow_share=("SCEN_OVERFLOW", "mean"),
               records=("SCEN_OCC_RATE", "count"),
           )
           .reset_index()
    )
    model_sum["scenario"] = label
    model_rows.append(model_sum)

system_summary = pd.DataFrame(system_rows)
sector_summary = pd.concat(sector_rows, ignore_index=True)
model_summary  = pd.concat(model_rows, ignore_index=True)

# Save outputs
system_summary.to_csv("outputs/scenario_system_summary.csv", index=False)
sector_summary.to_csv("outputs/scenario_sector_summary.csv", index=False)
model_summary.to_csv("outputs/scenario_program_model_summary.csv", index=False)

print("\nSaved scenario summaries:")
print("- outputs/scenario_system_summary.csv")
print("- outputs/scenario_sector_summary.csv")
print("- outputs/scenario_program_model_summary.csv")

print("\nSystem scenario summary:")
print(system_summary)

#-----Plot Scenario Results (System Level)-----


plt.figure(figsize=(10, 5))
plt.bar(system_summary["scenario"], system_summary["critical_share"])
plt.xticks(rotation=20, ha="right")
plt.title("System Critical Share (>=95% occupancy) by Scenario")
plt.xlabel("Scenario")
plt.ylabel("Critical Share")
plt.tight_layout()
plt.savefig("figures/model_03_scenario_critical_share.png", dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(system_summary["scenario"], system_summary["overflow_share"])
plt.xticks(rotation=20, ha="right")
plt.title("System Overflow Share (occupied > capacity) by Scenario")
plt.xlabel("Scenario")
plt.ylabel("Overflow Share")
plt.tight_layout()
plt.savefig("figures/model_04_scenario_overflow_share.png", dpi=300)
plt.show()

#-----Plot Sector Breakdown for a Key Scenario (Demand +10%)-----
# this will shows who gets hit hardest (equity + operations insight).
KEY_SCEN = "Demand +10%"
key_sector = sector_summary[sector_summary["scenario"] == KEY_SCEN].copy()
key_sector = key_sector.sort_values("critical_share", ascending=False)

plt.figure(figsize=(9, 5))
plt.barh(key_sector["SECTOR"], key_sector["critical_share"])
plt.gca().invert_yaxis()
plt.title(f"Critical Share by Sector — {KEY_SCEN}")
plt.xlabel("Critical Share (>=95%)")
plt.ylabel("Sector")
plt.tight_layout()
plt.savefig("figures/model_05_sector_critical_keyscenario.png", dpi=300)
plt.show()

#-----Buffer Policy Sweep (Find minimum capacity buffer for stability)-----
buffer_levels = np.arange(0.0, 0.26, 0.01)  # 0% to 25%
policy_rows = []

for b in buffer_levels:
    dsc = simulate(df, demand_increase=0.0, capacity_buffer=b)
    policy_rows.append({
        "capacity_buffer": b,
        "avg_occ_rate": float(np.nanmean(dsc["SCEN_OCC_RATE"])),
        "critical_share": float(np.mean(dsc["SCEN_CRITICAL_95"])),
        "overflow_share": float(np.mean(dsc["SCEN_OVERFLOW"])),
    })

policy_df = pd.DataFrame(policy_rows)
policy_df.to_csv("outputs/policy_buffer_sweep.csv", index=False)
print("\nSaved: outputs/policy_buffer_sweep.csv")

plt.figure(figsize=(10, 5))
plt.plot(policy_df["capacity_buffer"], policy_df["critical_share"], marker="o")
plt.title("Capacity Buffer vs System Critical Share (Baseline Demand)")
plt.xlabel("Capacity Buffer (fraction)")
plt.ylabel("Critical Share (>=95%)")
plt.tight_layout()
plt.savefig("figures/model_06_buffer_vs_critical.png", dpi=300)
plt.show()

#Choose a target critical share
TARGET = 0.10  # aim: <= 10% of daily records in critical state
eligible = policy_df[policy_df["critical_share"] <= TARGET]

if len(eligible) > 0:
    best_buffer = float(eligible.iloc[0]["capacity_buffer"])
    print(f"Minimum capacity buffer to reach critical_share <= {TARGET:.0%}: {best_buffer:.0%}")
else:
    print(f"No buffer up to 25% achieved critical_share <= {TARGET:.0%}. Consider increasing sweep max or revising target.")

#-----Produce Slide-Ready Key Numbers (printout)-----
baseline = system_summary[system_summary["scenario"] == "Baseline"].iloc[0]
d5 = system_summary[system_summary["scenario"] == "Demand +5%"].iloc[0]
d10 = system_summary[system_summary["scenario"] == "Demand +10%"].iloc[0]

print("\n===== SLIDE-READY KEY NUMBERS =====")
print(f"Baseline critical share: {baseline['critical_share']:.2%}")
print(f"Baseline overflow share:  {baseline['overflow_share']:.2%}")
print(f"Demand +5% critical share: {d5['critical_share']:.2%}  (Δ {d5['critical_share']-baseline['critical_share']:+.2%})")
print(f"Demand +10% critical share:{d10['critical_share']:.2%} (Δ {d10['critical_share']-baseline['critical_share']:+.2%})")
print("Phase 3 complete. Figures saved to 'figures/' and tables saved to 'outputs/'.")
```
